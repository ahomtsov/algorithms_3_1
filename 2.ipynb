```python
#Разработать алгоритм и программу интерполирующего поиска.
#В качестве исходных данных использовать массив целых чисел,
#который вводится с клавиатуры. Аргумент поиска – число.
```

```python
import numpy as np
import time as t
from random import randint
import plotly.express as px
import re
```

```python
def interpolation_search(searchable: list, number: int):
    low = 0
    high = len(searchable) - 1

    while number >= searchable[low] and number <= searchable[high]:
        #на каждой итерации значение уменьшается на Xpn, где p - mid = оценка
        mid = int(low + (number - searchable[low]) / (searchable[high] - searchable[low]) * (high - low))
        if searchable[mid] == number:
           # print(f"ключ {number} найден")
            return mid
        elif searchable[mid] > number:
            high = mid - 1
        elif searchable[mid] < number:
            low = mid + 1
        else:
            return -1
    
        
        
```

```python
 def time_count(searchable: list, number: int):
    start_time = t.time()
    interpolation_search(searchable, number)
    finis_time = t.time()
    return finis_time - start_time
```

```python
print("Ёмкостная сложность: O(n)")
print("Сложность алгоритма составляет O(loglogn)")
print(time_count([1,2,3,4,5,6,7,8], 8))
```

```python
def evaluate_results(repetitions: int) -> list:
    units = sorted([randint(1,1000) for _ in range(50_000_000)])
    meta = []
    for _ in range(repetitions):
        n = randint(0, len(units) - 1)
        evaluated_time = time_count(units, randint(1,1000))
        meta.append((n, evaluated_time))
    return meta
```

```python
def build_chart(raw):
    chart_data = []

    for (n, time) in raw:
        chart_data.append(
            dict(
                size=n,
                evaluation=time
            )
        )
    fig = px.line(chart_data, x="size", y="evaluation")
    fig.show()
```

```python

build_chart(sorted(evaluate_results(1000), key=lambda tup: tup[0]))
```
